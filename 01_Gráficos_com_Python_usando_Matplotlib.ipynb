```python
# importando as bibliotecas
import pandas as pd                 # manipulação de dados
import numpy as np                  # modelagem matemática

import matplotlib.pyplot as plt     # visualização gráfica
%matplotlib inline
```

**Orientação a _objetos_ com `Matplotlib`**

```python
# uma "figure" é uma janela
figura = plt.figure()
```

```python
# adicionando um gráfico na janela (figure)
grafico = figura.add_axes([0, 0, 1, 1])
```

```python
# visualizando a janela 
figura
```

```python
# criando uma série de dados
x = np.linspace(0, 100, 10)
y = x**3
```

```python
# visualizando a variável x
x
```

```python
# visualizando a variável y
y
```

```python
# gráfico de linha
grafico.plot(x, y, 'red');
```

```python
# visualizando o gráfico de linha
figura
```

**Gráfico de linha**

```python
# importando o conjunto de dados
df_bitcoin = pd.read_csv('https://datahub.io/cryptocurrency/bitcoin/r/bitcoin.csv')

# visualizando as 5 primeiras linhas
df_bitcoin.head()
```

```python
# dimensão do conjunto de dados
df_bitcoin.shape    # linhas e colunas
```

```python
# visualizando as 20 últimas linhas do atributo "price(USD)"
df_bitcoin['price(USD)'].tail(20)
```

```python
# instanciando e visualizando as 20 últimas linhas 
df_bitcoin = df_bitcoin.tail(20)
df_bitcoin
```

```python
# plotando gráfico de linhas da seleção das 20 últimas linhas do "df_bitcoin"
fig = plt.figure(figsize=(10, 5))
graf = fig.add_axes([0, 0, 1, 1])
graf.plot(df_bitcoin['date'], df_bitcoin['price(USD)'], 'green')
plt.xticks(rotation=45);
```

**Gráfico de barras**

```python
# importando e visualizando as 5 primeira linhas do conjunto de dados
df_titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df_titanic.head()
```

```python
# nome dos atributos
df_titanic.columns
```

```python
# realizando uma seleção dos atributos de interesse, para isso, utilizaremosa função ".groupby()" 
# e contabilizando  suas instâncias presente
sexo = df_titanic.groupby('Sex')['Survived'].sum().reset_index()
sexo
```

```python
# plotando gráfico de barras da seleção realizada acima
fig = plt.figure(figsize=(10, 5))
graf = fig.add_axes([0, 0, 1, 1])
graf.bar(sexo['Sex'], sexo['Survived']);
```

**Histograma**

```python
# plotando gráfico de histograma da distribuição do atributo "Age"
fig = plt.figure(figsize=(10, 5))
graf = fig.add_axes([0, 0, 1, 1])
graf.hist(df_titanic['Age'], bins=25);
```

**Boxplot**

```python
# visualizando o tipo de dado dos atributos
df_titanic.dtypes
```

```python
# contabilizando a presença de valores nulos
df_titanic.isnull().sum()
```

```python
# removendo os valores nulos e instanciando o resultado da operação na variável "age"
age = df_titanic.dropna()
```

```python
# visualizando se o tratamento ocorreu com sucesso
age['Age'].isnull().sum()
```

```python
# estatística descritiva do conjunto de dados após a remoção de valores nulos
age.describe()
```

```python
# plotando gráfico de boxplot do atributo "Age"
fig = plt.figure(figsize=(10, 5))
graf = fig.add_axes([0, 0, 1, 1])
graf.boxplot(age['Age']);
```

```python
# instanciando valor discrepante para visualizarmos no boxplot
idade_outlier = list(age['Age'])
idade_outlier.append(125)
```

```python
# plotando gráfico de boxplot do atributo "Age" com o valor discrepante (outlier)
fig = plt.figure(figsize=(10, 5))
graf = fig.add_axes([0, 0, 1, 1])
graf.boxplot(idade_outlier);
```

**Criando vários gráficos em grade**

```python
# criando um gráfico 
fig, graf = plt.subplots(nrows=1, ncols=1)
```

```python
# removendo o gráfico da janela (figure)
graf.remove()
fig
```

```python
# criando um gráfico
fig, graf = plt.subplots(nrows=1, ncols=1)
```

```python
# criando um gráfico e definindo o seu tamanho de 5 x 5
fig, graf = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
```

```python
# criando um gráfico e definindo o seu tamanho de 10 x 5
fig, graf = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
```

```python
# criando um gráfico e definindo o seu tamanho de 5 x 10
fig, graf = plt.subplots(nrows=1, ncols=1, figsize=(5, 10))
```

```python
# criando quatro gráficos e definindo tamanho de 10 x 5 cada
fig, graf = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
```

```python
# criando quatro gráficos e definindo tamanho de 10 x 5 cada 
# e com a função ".tight_layout()" melhorando sua visualização
fig, graf = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
plt.tight_layout()
```
