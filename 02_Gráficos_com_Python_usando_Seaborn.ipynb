**Gráfico de linha**

```python
# importando o conjunto de dados
df_bitcoin = pd.read_csv('https://datahub.io/cryptocurrency/bitcoin/r/bitcoin.csv')

# visualizando as 5 primeiras linhas
df_bitcoin.head()
```

```python
# dimensão do conjunto de dados
df_bitcoin.shape    # linhas e colunas
```

```python
# visualizando as 20 últimas linhas do atributo "price(USD)"
df_bitcoin['price(USD)'].tail(20)
```

```python
# instanciando e visualizando as 20 últimas linhas 
df_bitcoin = df_bitcoin.tail(20)
df_bitcoin
```

```python
plt.figure(figsize=(12, 7))
plt.xticks(rotation=45)
sns.set_theme(style='darkgrid')
sns.lineplot(data=df_bitcoin, x='date', y='price(USD)', color='orange');
```

**Gráfico de barras**

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sobreviventes_por_sexo = titanic.groupby('sex')['survived'].sum().reset_index()
sobreviventes_por_sexo
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=sobreviventes_por_sexo, x='sex', y='survived');
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=titanic, x='sex');
```

**Histograma**

```python
plt.figure(figsize=(12, 7))
sns.histplot(data=titanic, x='age', binwidth=5);
```

**Boxplot**

```python
titanic.dtypes
```

```python
titanic.isnull().sum()
```

```python
df_titanic = titanic.dropna()
```

```python
df_titanic.isnull().sum()
```

```python
df_titanic['age'].isnull().sum().any()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=df_titanic, x='age');
```

```python
plt.figure(figsize=(12, 7))
sns.boxenplot(data=df_titanic, x='age', y='sex');
```

```python
plt.figure(figsize=(12, 7))
sns.boxenplot(data=df_titanic, x='age', y='class');
```

```python
plt.figure(figsize=(12, 7))
sns.boxenplot(data=df_titanic, x='age', y='sex', hue='survived');
```

```python
# instanciando valor discrepante para visualizarmos no boxplot
idade_outlier = list(df_titanic['age'])
idade_outlier.append(125)
idade_outlier
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=idade_outlier, orient='h');
```

**Mapas de calor (heatmap)**

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
df_flights = flights.pivot('month', 'year', 'passengers')
df_flights
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(data=df_flights, annot=True, fmt='.0f');
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(data=df_flights, annot=True, fmt='.0f', cmap='YlGnBu');
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(data=df_flights, annot=True, fmt='.0f', cmap='Blues');
```

**Gráficos de dispersão com regressão linear**

```python
accidents =  sns.load_dataset('car_crashes')
accidents.head()
```

**Seaborn com Matplotlib**

```python
fig, grafs = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
plt.tight_layout();
```

```python
sns.barplot(data=sobreviventes_por_sexo, x='sex', y='survived', ax=grafs[0][0])
sns.histplot(data=df_titanic, x='age', ax=grafs[0][1])
sns.boxplot(data=idade_outlier, orient='h', ax=grafs[1][0])
sns.heatmap(df_flights, annot=True, fmt='.0f', cmap='YlGnBu', ax=grafs[1][1])

fig
```
